# Algoritmos de extraccion de texto de hmtl

```python
from bs4 import BeautifulSoup

ext='.html'
dir_local='2021/html/Jurisprudencia Nacional Sistematizada'
list_of_list_page_webs_character=[]
list_of_list_page_webs_url=[]
list_of_list_page_webs_html=[]
for id_html in range(315):
    
    # Opening the html file
    try:
        print("entrada : ",dir_local+str(id_html)+ext)
        HTMLFile = open(dir_local+str(id_html)+ext, "r")

        # Reading the file
        index = HTMLFile.read()

        # Creating a BeautifulSoup object and specifying the parser
        S = BeautifulSoup(index, 'lxml')

        # Using the find_all method to find all elements of a tag
        festures_pdf=[]
        for tag in S.find_all('div'):

          # Printing the name, and text of p tag
            #print(f'{tag.name}: {tag.text}')
            string_with_nonASCII=tag.text
            festures_pdf.append(string_with_nonASCII)
        
        festures_pdf2=[]
        for j in festures_pdf:
            text = ''.join(j)
            text=(text.replace('\n \n',', ').replace('\n',' '))
            text=text.replace(u'\xa0', u' ')
            text=text.replace(u'\t', u' ')
            text = text.split(",")
            text2 = ''.join(text)
            text2=" ".join(text2.split())
            #text2 = text2.split(" ")

            festures_pdf2.append(text2)
        ###################################################################3
        new_text_doc=' '.join(festures_pdf2)
        new_text_doc=new_text_doc.split(" ")
        #new_text_doc
        ###################################################################33333
        #Detectamos apelacion, demanda, queja de un archivo html y se crear un lista de ubicacion de cada palabra clave detectado
        id_apelacion=[]
        for m in range(len(new_text_doc)):
            if new_text_doc[m]=="Apelacin" and new_text_doc[m+2]=="Especialidad:" and new_text_doc[m+3]=="Laboral":
                #print(new_text_doc[m], new_text_doc[m+1], new_text_doc[m+2])
                id_apelacion.append(m)
            if new_text_doc[m]=="Consulta" and new_text_doc[m+1]=="/" and new_text_doc[m+2]=="Revisin" and new_text_doc[m+4]=="Especialidad:":
                #print(new_text_doc[m], new_text_doc[m+1], new_text_doc[m+2])
                id_apelacion.append(m)
            if new_text_doc[m]=="Queja" and new_text_doc[m+2]=="Especialidad:" and new_text_doc[m+3]=="Laboral":
                #print(new_text_doc[m], new_text_doc[m+1], new_text_doc[m+2])
                id_apelacion.append(m)
            if new_text_doc[m]=="Demanda" and new_text_doc[m+2]=="Especialidad:" and new_text_doc[m+3]=="Laboral":
                #print(new_text_doc[m], new_text_doc[m+1], new_text_doc[m+2])
                id_apelacion.append(m)
                
                
        ####################################################################
        text_features_doc=[]
        for n in range(len(id_apelacion)):
            #print(id_apelacion[n])
            #print(new_text_doc[id_apelacion[n]])

            if n < len(id_apelacion)-1:
                #print(new_text_doc[id_apelacion[n]:id_apelacion[n+1]])
                text_features_doc.append(new_text_doc[id_apelacion[n]:id_apelacion[n+1]])
            else:
                text_features_doc.append(new_text_doc[id_apelacion[n]: ])
        ############################################################################
        #Se estructura informacion de cada apelacion, demanda, queja ... se crea lista de ubicacion de cada palabra clave
        df_list_p=[]
        for p in range(10):
            df_list_h=[]
            for s in range(len(text_features_doc[p])):
                if text_features_doc[p][s]=="Apelacin" or text_features_doc[p][s]=="Queja" or text_features_doc[p][s]=="Consulta" or text_features_doc[p][s]=="Demanda":
                    if text_features_doc[p][s]=="Apelacin" or text_features_doc[p][s]=="Queja" or text_features_doc[p][s]=="Demanda":
                        if text_features_doc[p][s+2]=="Especialidad:" and text_features_doc[p][s+3]=="Laboral":
                            df_list_h.append(s)
                    if text_features_doc[p][s]=="Consulta" and text_features_doc[p][s+1]=="/" and text_features_doc[p][s+2]=="Revisin" and text_features_doc[p][s+4]=="Especialidad:":
                        df_list_h.append(s)
                        
                if text_features_doc[p][s]=="Especialidad:":
                    df_list_h.append(s)
                if text_features_doc[p][s]=="Resolucin:"  and text_features_doc[p][s-1]=="Tipo":
                    df_list_h.append(s-1)
                if text_features_doc[p][s]=="Resolucin:" and text_features_doc[p][s-1]=="Fecha":
                    df_list_h.append(s-1)
                if text_features_doc[p][s]=="Jurisdiccional:" and text_features_doc[p][s-1]=="rgano":
                    df_list_h.append(s-1)
                if text_features_doc[p][s]=="Materia:":
                    df_list_h.append(s)
                if text_features_doc[p][s]=="Sumilla:":
                    df_list_h.append(s)
                if text_features_doc[p][s]=="Clave:" and text_features_doc[p][s-1]=="Palabras":
                    df_list_h.append(s-1)
            df_list_p.append(df_list_h)
        ####################################################################################3
        #se ejecuta la lista de ubicacion de palabras claves del anterior paso para concatenar y estructurar
        df_list_p1=[]
        for p in range(10):
            df_list_h1=[]
            #print(len(df_list_p[p]))
            for s in range(len(df_list_p[p])):

                #print(len(df_list_p[p]))

                #print(df_list_p[p][s])

                #print(s)
                if s < 7:
                    #print(text_features_doc[p][df_list_p[p][s]:df_list_p[p][s+1]])

                    text=text_features_doc[p][df_list_p[p][s]:df_list_p[p][s+1]]
                    text=' '.join(text)
                    df_list_h1.append(text)
                else:
                    #print(text_features_doc[p][df_list_p[p][s]: ])

                    text=text_features_doc[p][df_list_p[p][s]: ]
                    text=' '.join(text)
                    df_list_h1.append(text)

            df_list_p1.append(df_list_h1)
            list_of_list_page_webs_character.append(df_list_h1)
            print(len(df_list_h1))
        #################################################################################################33
        #################################################################################################33
        
  
        # Creamos la lista de urls de las resoluciones 
        S = BeautifulSoup(index, 'lxml')

        links = S.find_all('a')
        urlss=[]
        for tag in links:
            link = tag.get('href',None)
            if len(link)==108:
                #print(dir_local+str(id_html)+ext, link)
                urlss.append(link)
                list_of_list_page_webs_url.append(link)
                list_of_list_page_webs_html.append(dir_local+str(id_html)+ext)
            
    except:
        print(dir_local+str(id_html)+ext, )
        pass
```

```python
import pandas as pd
df_fetures = pd.DataFrame(list_of_list_page_webs_character, columns = ['Apelacin', 'Especialidad','Tipo_Resolucin','Fecha_Resolucin','rgano_Jurisdiccional','Materia','Sumilla','Palabras_Clave'])
```

# Descarga de documentos pDF

```python
import urllib.request
import uuid

def download_file(download_url, filename):
    try:
        response = urllib.request.urlopen(download_url)    
        file = open(filename + ".pdf", 'wb')
        print(file)
        file.write(response.read())
        file.close()
    except:
        pass 
```
